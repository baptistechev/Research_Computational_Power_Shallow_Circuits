```python
import graphix as graph
```

```python
#GHZ State (on qubit 0,3,4)

#Start with 3-star graph
pat = graph.pattern.Pattern([0,1,2])
pat.add(graph.command.E(nodes=(0,1)))
pat.add(graph.command.E(nodes=(0,2)))

#Apply Hadamard to qubits 1,2
pat.add(graph.command.N(node=3))
pat.add(graph.command.E(nodes=(1,3)))
pat.add(graph.command.M(node=1,angle=0))
pat.add(graph.command.X(node=3,domain=[1]))

pat.add(graph.command.N(node=4))
pat.add(graph.command.E(nodes=(2,4)))
pat.add(graph.command.M(node=2,angle=0))
pat.add(graph.command.X(node=4,domain=[2]))

pat.print_pattern()
pat.draw_graph()
```

```python
state = pat.simulate_pattern(backend='statevector')
state
```

```python
gs = graph.GraphState(nodes=[1,2,3,4,5,6],edges=[(1,2),(2,3),(4,5),(5,6),(1,4),(2,5),(3,6)])
gs.measure_x(1,0)
gs.equivalent_graph_E2(2,4)
gs.measure_y(2,0)
gs.measure_z(4,0)
gs.draw()
```

```python
gs = graph.GraphState(nodes=[1,2,3,4],edges=[(1,2),(2,3),(3,4),(4,1)])
gs.z(2)
gs.s(3)
gs.h(3)
gs.h(4)
gs.equivalent_graph_E1(3)
gs.draw()
```
